# **History of Python**

##**-Version 0.90**  

The codigode of that version was published by Van Rossum in 1991, classes with inheritance, exception handling, functions and modular types were already available in that version. A few explaces are str, list, dict, etc.

In that release, a system of adopted modules appeared from Modula-3, a structured and modular programming language, which Guido describes as one of the largest Python programming units.
By 1994 comp.lang.python was created, a Python discussion forum that marked a milestone in its popularity and multiplied its number of users.

## **-Version 1.0**

This version is published in 1994 with some programming tools added such as lambda, reduce, filter and map. And these additions were thanks to a hacker from Lisp, a family of multiparadigm computer programming languages.
The last version released still belonging to the research center was version 1.2, from this moment Van Rossum decided to continue his project freely, after this the next version released was version 1.4 and this included new features based on modula-3, in addition built-in support for complex numbers.
By 2000's, the core team of Python developers switched to BeOpen.com to form the BeOpen Python Labs team.
The next version and last of the first generation was version 1.6 of Python had some problems with its license type until the Free Software Foundation (FSF) managed to change Python to a Free Software license, which would make it GPL-compatible.

**Interesting fact:** During his stay at CNRI, Van Rossum launched the Computer Programming for Everybody initiative (CP4E), to make programming more accessible to more people, with a basic 'literacy' level in programming languages, similar to the basic English and maths skills needed by many employers. Python played a crucial role in this process: Because of its orientation toward a clean syntax, it was already ideal and CP4E's goals bore similarities to its predecessor, ABC. The project was sponsored by DARPA. But unfortunately this CP4E project is currently inactive.


## **-Version 2.0**

That version was published in October 2000, including list generation, one of the most important features of the Haskell functional programming language. It also included a garbage collection system capable of collecting cyclical benchmarks.
In 2001, the Python Software Foundation was created, which since Python 2.1 owns all the code, documentation, and language specifications. The foundation was based on the Apache Software Foundation model. That release included a change in the language to support nested scopes (better known in programming as "nested scopes") as supported by other "static scoping" languages.
In the next update Python 2.2 was the unification of types in Python (types written in C), and classes (types written in Python) within a hierarchy. This achieved a pure and consistent Python object-oriented model, and generators were added that were inspired by the Icon language.

### **Else Statement**

The else command is executed when an if command was not executed, to use it we must write an if command and then an else command. Example:

```python
f=20
g=40
if f > g:
  print("I'm over 40 years old.")
else: 
  print("I am under 40 years old")
```

### **Elif Statement**

The elif command is used to add an extra condition between the if command and the else command. Example:

```python
f=20
g=40
if f > g:
  print("I'm over 40 years old.")
elif  f < g:
  print("I am under 40 years old")
else: 
  print("I am between 20 and 40 years old")
```
